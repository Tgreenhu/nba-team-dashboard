```python
# Import Dependencies

import pandas as pd
from bs4 import BeautifulSoup as soup
from splinter import Browser
import matplotlib.pyplot as plt
```

```python
# Set Path To Dataset

browser = Browser('chrome')
url = 'https://www.basketball-reference.com/leagues/NBA_2019_advanced.html'
browser.visit(url)
```

```python
# Create BS object

html = browser.html
soup_obj = soup(html, 'html.parser')
```

```python
# Scrape 2018-2019 stats

table = soup_obj.find_all('table', class_='stats_table')
table_df = pd.read_html(str(table))[0]
table_df.head()
```

```python
# Delete columns we will not need

del table_df['Rk']
del table_df['Pos']
del table_df['G']
del table_df['MP']
del table_df['TS%']
del table_df['3PAr']
del table_df['FTr']
del table_df['ORB%']
del table_df['DRB%']
del table_df['TRB%']
del table_df['AST%']
del table_df['STL%']
del table_df['BLK%']
del table_df['TOV%']
del table_df['OWS']
del table_df['DWS']
del table_df['WS']
del table_df['WS/48']
del table_df['OBPM']
del table_df['DBPM']
del table_df['BPM']
del table_df['Unnamed: 19']
del table_df['Unnamed: 24']
table_df.head()
```

```python
# Scatter Plot USG% vs VORP

x_axis = table_df['USG%']
y_axis = table_df['VORP']
plt.scatter(x_axis, y_axis);
```
