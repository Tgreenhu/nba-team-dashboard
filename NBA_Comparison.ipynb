```python
# Import Dependencies

import sqlalchemy
from sqlalchemy import create_engine
import requests
import pandas as pd
from bs4 import BeautifulSoup as soup
from splinter import Browser
```

```python
# Set Path To Dataset

browser = Browser('chrome')
url = 'https://www.basketball-reference.com/leagues/NBA_2019_advanced.html'
browser.visit(url)
```

```python
# Create BS object

html = browser.html
soup_obj = soup(html, 'html.parser')
```

```python
# Scrape 2018-2019 stats

table = soup_obj.find_all('table', class_='stats_table')
table_df = pd.read_html(str(table))[0]
table_df.head()
```

```python
new_df = table_df[['Player', 'Age', 'Tm', 'MP', 'PER', 'TOV%', 'USG%', 'VORP']]
new_df.head(25)
```

```python
nba_df = new_df.drop(new_df[new_df['Player'] == 'Player'].index)
nba_df.head(25)
```

```python
new_nba_df = nba_df.drop(nba_df[nba_df['Tm'] == 'TOT'].index)
new_nba_df.head(10)
```

```python
clean_df = new_nba_df.rename(columns={
    'Player': 'name',
    'Age': 'age',
    'Tm': 'team',
    'MP': 'minutes',
    'PER': 'per',
    'TOV%': 'turnover',
    'USG%': 'usg',
    'VORP': 'vorp'
})
clean_df.head(10)
```

```python
# Connect To Postgres

connection_string = f"postgres:Georgia-07@localhost:5432/nba_comp"
engine = create_engine(f'postgresql://{connection_string}')
conn = engine.connect()
```

```python
# Insert DF Into Postgres

clean_df.to_sql('nba', con = engine, if_exists = 'append', index=False)
```

```python
# Cheack Connection From Postgres to Pandas
data = pd.read_sql('SELECT * FROM nba', conn)
data.head(20)
```
